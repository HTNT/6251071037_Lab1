```python
import pandas as pd
import matplotlib.pyplot as plt
from openpyxl import Workbook
```

```python
#a
df = pd.read_excel("Dataset/tb_tracking.xlsx", header=None)
df.columns = ["id", "device_id", "co_level", "humidity", "temperature", "time"]
print(df)
```

```python
#b
df["time"] = pd.to_datetime(df["time"], errors='coerce')
df = df[df["time"] >= pd.to_datetime("2018-01-03")]
```

```python
#c
df["minutes"] = df["time"].dt.hour * 60 + df["time"].dt.minute
```

```python
#d
# CO LEVEL
plt.figure(figsize=(10,5))
plt.scatter(df["minutes"], df["co_level"])
plt.xlabel("Minutes")
plt.ylabel("CO Level")
plt.title("CO Level vs Minutes")
plt.show()

# HUMIDITY
plt.figure(figsize=(10,5))
plt.scatter(df["minutes"], df["humidity"])
plt.xlabel("Minutes")
plt.ylabel("Humidity")
plt.title("Humidity vs Minutes")
plt.show()

# TEMPERATURE
plt.figure(figsize=(10,5))
plt.scatter(df["minutes"], df["temperature"])
plt.xlabel("Minutes")
plt.ylabel("Temperature")
plt.title("Temperature vs Minutes")
plt.show()

print(df.head())
```
